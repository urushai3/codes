```python
import requests
from bs4 import BeautifulSoup

# Define the base URL and query for PubMed page
base_url = "https://pubmed.ncbi.nlm.nih.gov/?term=Carlos+H+Barcenas&page="

# Initialize lists to store publications and citations
all_publications = []
all_citations = []

# Loop through pages
for page_num in range(1, 7):
    # Construct the URL for the current page
    current_url = base_url + str(page_num)

    # Send a GET request
    response = requests.get(current_url)

    # Parse the content using BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Extract relevant information
    publications = soup.find_all("a", {"class": "docsum-title"})
    citations = soup.find_all("span", {"class": "docsum-journal-citation full-journal-citation"})

    # Append publications and citations to respective lists
    for publication in publications:
        all_publications.append(publication.text.strip())

    for citation in citations:
        all_citations.append(citation.text.strip())

# Print publications and citations with their count
for idx, (publication, citation) in enumerate(zip(all_publications, all_citations), start=1):
    print(f"{idx}. Publication: {publication}\n   Citation: {citation}\n")
```

```python
 import requests
from bs4 import BeautifulSoup

# Define the base URL and query for PubMed page
base_url = "https://pubmed.ncbi.nlm.nih.gov/?term=Carlos+H+Barcenas&page="

# Initialize lists to store publications, citations, and co-authors
all_publications = []
all_citations = []
all_coauthors = []

# Loop through pages
for page_num in range(1, 7):
    # Construct the URL for the current page
    current_url = base_url + str(page_num)

    # Send a GET request
    response = requests.get(current_url)

    # Parse the content using BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Extract relevant information
    publications = soup.find_all("a", {"class": "docsum-title"})
    citations = soup.find_all("span", {"class": "docsum-journal-citation full-journal-citation"})
    coauthors = soup.find_all("span", {"class": "docsum-authors full-authors"})

    # Append publications, citations, and co-authors to respective lists
    for publication in publications:
        all_publications.append(publication.text.strip())

    for citation in citations:
        all_citations.append(citation.text.strip())

    for coauthor in coauthors:
        all_coauthors.append(coauthor.text.strip())

# Print publications, citations, and co-authors with their count
for idx, (publication, citation, coauthor) in enumerate(zip(all_publications, all_citations, all_coauthors), start=1):
    print(f"{idx}. Publication: {publication}\n   Citation details: {citation}\n   Co-authors: {coauthor}\n")
```

```python
import requests
from bs4 import BeautifulSoup
import json

def get_publication_info_for_pmid(pmid, counter):
    url = f'https://pubmed.ncbi.nlm.nih.gov/{pmid}/'

    # Fetch HTML content from the provided URL
    html_text = requests.get(url).text
    # Create BeautifulSoup object
    soup = BeautifulSoup(html_text, 'lxml')

    # Find the publication name
    publication_name = soup.find('h1', class_='heading-title').text.strip()

    # Find the container with co-author information
    co_authors_container = soup.find('div', class_='authors-list')
    if co_authors_container is None:
        print(f"Co-authors not found for PMID {pmid}.")
        return

    # Initialize a dictionary to store publication information
    Affiliation_data = {
        "PMID": pmid,
        "Publication": publication_name,
        "Co-Authors": []
    }

    # Find all co-authors
    author_items = co_authors_container.find_all('span', class_='authors-list-item')

    # Find all affiliation links
    affiliation_links = soup.find_all('a', class_='affiliation-link')

    # Iterate through co-authors and affiliations
    for author_item, affiliation_link in zip(author_items, affiliation_links):
        # Extract author name
        author_name = author_item.find('a', class_='full-name').text.strip()

        # Extract affiliation
        affiliation = affiliation_link['title']

        # Store co-author and affiliation data in the dictionary
        Affiliation_data["Co-Authors"].append({
            "Author": author_name,
            "Affiliation": affiliation
        })

    # Convert the publication data to JSON
    json_data = json.dumps(Affiliation_data, indent=4)
    print(json_data)
    print('\n')

# List of PMIDs
all_pmids = ['35175860', '28928852', '32219153', '26402167', '31892799', '34351787', '35379811', '35326615', '37921773', '32253074', '34710521', '33678567', '35816269', '31461380', '31263363', '32431013', '29752326', '37733286', '28742682', '29689595', '30795923', '34247335', '33691275', '28135395', '37973647', '33922702', '28957377', '31097774', '16421416', '28844850', '32214335', '22371383', '37735019', '35588361', '30139836', '28085940', '24868022', '36702070', '36863241', '36681001', '28376179', '28199747', '26979306', '28720217', '29928946', '30077429', '28766219', '31184783', '28423171', '29239836', '30977027', '15776474', '17426341', '26678405', '15914831', '17925478', '24600161', '17103253', '20133260']  # Replace this with your list of PMIDs

all_json_data = []

# Iterate through the list of PMIDs
for counter, pmid in enumerate(all_pmids, start=1):
    publication_json = get_publication_info_for_pmid(pmid, counter)
    if publication_json:
        all_json_data.append(publication_json)

# Write all JSON data to a file
with open('Affiliation_data.json', 'w') as json_file:
    for json_data in all_json_data:
        json_file.write(json_data + '\n')  #
```
